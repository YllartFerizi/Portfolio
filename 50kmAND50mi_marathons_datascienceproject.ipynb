# Data Science Project :50km and 50mi races in USA

### Importing the data

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
dataset.head()
```

### Cleaning Up data
**We only are interested on USA races,50km ore 50miles that happened on 2020**

```python
dataset[dataset["Event distance/length"].isin(['50km','50mi'])]
```

```python
dataset[dataset['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)=='USA']
```

```python
dataset2 = dataset[(dataset["Event distance/length"].isin(['50km','50mi'])) & (dataset["Year of event"]==2020) & (dataset['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)=='USA')]
```

```python
dataset2.info()
```

```python
dataset2.shape
```

```python
dataset2.head(5)
```

```python
dataset2["Event name"]=dataset2["Event name"].str.split('(').str.get(0)
```

```python
dataset2.head()
```

```python
dataset2["Athlete performance"]=dataset2["Athlete performance"].str.split(' ').str.get(0)
```

```python
dataset2['Athlete Age']=2020 - dataset2['Athlete year of birth']
```

```python
dataset2.drop(['Athlete year of birth',"Athlete club","Athlete country","Athlete age category"],axis = 1,inplace=True)
```

```python
dataset2.head()
```

```python
dataset2.isna().sum()
```

```python
dataset2[dataset2['Athlete Age'].isna()==1]
```

```python
dataset2 = dataset2.dropna()
```

```python
dataset2.shape
```

```python
dataset2[dataset2.duplicated()==True]
```

```python
dataset2.reset_index(drop=True,inplace=True)
```

```python
dataset2.dtypes
```

```python
dataset2["Athlete Age"]=dataset2["Athlete Age"].astype(int)
```

```python
dataset2["Athlete average speed"]=dataset2["Athlete average speed"].astype(float)
```

```python
dataset2.dtypes
```

```python
dataset2=dataset2.rename(columns ={
    'Year of event': "year",
     "Event dates" : "dates",
     "Event name" : "event_name",
    "Event distance/length": "race_length",
    "Event number of finishers": "race_number_of_finisher",
    "Athlete performance":"athlete_performance" ,
    "Athlete gender": "athlete_gender",
    "Athlete average speed" :"athlete_average_soeed" ,
    "Athlete ID": "athlete_id", 
    "Athlete Age": "athlete_age",



    
})
```

```python
dataset2.head()
```

```python
dataset2.head()
```

## Plotting the Data

##### Number of Races ran in 50km and number of races ran in 50mi

```python
kmandmiles = [len(dataset2[dataset2["race_length"]=='50km']),len(dataset2[dataset2["race_length"]=='50mi'])]
Labels = "Number of Races ran in 50 km","Number of Races ran in 50 km"
plt.pie(kmandmiles,labels=Labels,autopct='%1.1f%%',explode=[0.08,0.01])
plt.show()
```

##### Races ran by a male vs Races ran by a female

```python
genders = [len(dataset2[dataset2["athlete_gender"]=='M']),len(dataset2[dataset2["athlete_gender"]=='F'])]
Labels = "Races ran by a Male","Races ran by a Female"
plt.pie(genders,labels=Labels,autopct='%1.1f%%',explode=[0.08,0.01])
plt.show()
```

##### Distribution Plot of average athtlete speed in the 50 mi race

##### Distribution Plot of average athtlete speed in the 50 km race

#### splitting the data in 4 different seasons

```python
dataset2["race_month"]= dataset2['dates'].str.split('.').str.get(1).astype(int)
```

```python
dataset2.head()
```

```python
dataset2["race_season"] = dataset2["race_month"].apply(lambda x :'Winter' if x >11 else 'Fall' if x> 8 else 'Summer' if x >5 else 'Spring'if x > 2 else 'Winter')
```

```python
dataset2.head()
```

```python
seasons = [len(dataset2[dataset2["race_season"]=='Winter']),len(dataset2[dataset2["race_season"]=='Fall']),len(dataset2[dataset2["race_season"]=='Summer']),len(dataset2[dataset2["race_season"]=='Spring'])]
Labels =["Races ran in Winter","Races ran in Fall","Races ran in Summer","Races ran in Spring"]
plt.pie(seasons,labels=Labels,autopct='%1.1f%%')
plt.show()
```

#### Average and the numbers of races ran in each season

```python
dataset2.groupby("race_season")['athlete_average_soeed'].agg(['mean','count']).sort_values(by='mean',ascending=False)
```

#### 50 miles only

```python
dataset2.query('race_length=="50mi"').groupby("race_season")['athlete_average_soeed'].agg(['mean','count']).sort_values(by='mean',ascending=False)
```
